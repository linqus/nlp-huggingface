```python
from transformers import pipeline
```

```python
sentiment = pipeline('sentiment-analysis')
```

```python
sentiment('Strawberries make me goosebumps')
```

```python
classifier = pipeline('zero-shot-classification')
```

```python
classifier('Spending time in this course will make you more interesting person', candidate_labels=['education', 'travells', 'social skills'])
```

```python
generator = pipeline('text-generation', model='distilgpt2')
```

```python
generator("In this course i'll teach you how to", num_return_sequences=2, max_length=30)
```

```python
fill = pipeline('fill-mask')
```

```python
fill('This course will teach you all about <mask> models.', top_k=3)
```

```python
fill = pipeline('fill-mask', model="bert-base-cased")
```

```python
ner = pipeline('ner', grouped_entities=True)
```

```python
ner('My name is Sylvain, I work in Hugging Face and I live in Brooklyn')
```

```python
ner = pipeline('ner', grouped_entities=True, model='QCRI/bert-base-multilingual-cased-pos-english')
```

```python
ner('My name is Sylvain, I work in Hugging Face and I live in Brooklyn')
```

```python
translation = pipeline('translation_en_to_de')
```

```python
translation('I work in Huging Face')
```
